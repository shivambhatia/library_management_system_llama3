```python
import requests
import random

# The URL to your Flask application's add-review endpoint
url = 'http://127.0.0.1:5000/add-review'

# Data for the new review you want to add

for i in data:
        ratings=df_ratings[df_ratings['book_id']==i["id"]]["rating"]
        for j in ratings:
            if j == 1:
                new_comment = random.choice(negative_samples)
            elif j in [2, 3]:
                new_comment = random.choice(neutral_samples)
            elif j in [4, 5]:
                new_comment = random.choice(positive_samples)
        
            review_data = {
                'book_id': i["id"],  # Assuming the book ID you're reviewing exists
                'user_id': 1,  # Assuming the user ID exists
                'rating': j,  # Rating out of 5
                'review_text': new_comment
            }
            # Sending a POST request to the endpoint with the review data
            response = requests.post(url, json=review_data)

            # Checking the response from the server
            if response.status_code == 201:
                print('Review added successfully.')
            else:
                print('Failed to add the review. Status code:', response.status_code)
        


```
